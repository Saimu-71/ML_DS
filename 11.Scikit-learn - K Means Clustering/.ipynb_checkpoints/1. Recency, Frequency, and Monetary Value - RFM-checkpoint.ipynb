## Recency

```python
currentDate = pd.to_datetime("2012-01-01")
currentDate
```

**Convert Date Formate**

```python
df['InvoiceDate']=pd.to_datetime(df["InvoiceDate"])
df['InvoiceDate'].head(2)
```

```python
df
```

```python
currentDate
```

```python
df.dtypes
```

```python
maxdate = df.groupby(["CustomerID"]).max()[['InvoiceDate']]
maxdate
```

```python
(currentDate - maxdate['InvoiceDate']).dt.days
```

```python
maxdate['InvoiceAge'] = (currentDate - maxdate['InvoiceDate']).dt.days
maxdate
```

```python
# Recency
recency = maxdate.drop('InvoiceDate', axis=1)
recency
```

## Frequency

```python
freq=df.drop_duplicates(subset="InvoiceNo")
freq
```

```python
# Frequency
freq = freq.groupby(["CustomerID"]).count()[["InvoiceNo"]]
freq
```

## Monetary

```python
df["total"]=df["Quantity"]*df["UnitPrice"]
```

```python
df
```

```python
# Monetary
money=df.groupby(["CustomerID"])[["total"]].sum()
money
```

```python
df[df['CustomerID']==12346.0]
```

```python
RFM=pd.concat([recency,freq,money], axis=1)
RFM
```

```python
RFM.columns = ['Recency', 'Frequency', 'Monetary']
RFM
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled=scaler.fit_transform(RFM)
```

```python
scaled
```

```python
scaled.shape
```

```python
from sklearn.cluster import KMeans
```

### To find out the value of K

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(scaled)
    sse.append(km.inertia_)
```

```python
sse
```

**Elbo Method, so I get k=3 is best**

```python
plt.plot(k_range, sse, 'r-*')
plt.xlabel ='k'
plt.ylabel = 'Sum of Squared Error'
```

### Object Creation and data fitting

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(scaled)
```

```python
kmeans.labels_
```

```python
kmeans.labels_.shape
```

```python
RFM["Clusters"]=(kmeans.labels_)
```

```python
RFM
```

```python
RFM.Clusters.unique()
```

```python
RFM[RFM.Clusters==0]
```

```python
RFM[RFM.Clusters==1]
```

```python
RFM[RFM.Clusters==2]
```

```python
final=RFM.groupby("Clusters").mean()[["Recency","Frequency","Monetary"]]
final
```

```python
def func(row):
    if row["Clusters"]==2:
        return 'Dimond'
    elif row["Clusters"]==0:
        return 'Gold'
    else:
        return 'Silver'
```

```python
RFM['group']=RFM.apply(func, axis=1)
```

```python
RFM
```

```python
result = pd.DataFrame(RFM.group.value_counts())
result
```

```python
result = result.reset_index()
result
```

```python
result.plot(kind="bar", color=["Orange","Red","Green"])
```
