**5. What are the three most common customer's names?**<br>

\**6. Two customers have the same phone number, can you find those customers**?

```python
phone_counts = df['phone'].value_counts()
same_phone = phone_counts[phone_counts == 2]

for phone, count in same_phone.items():
    print(f"{phone}    {count}")
```

**7. How many customers have profession "Structural Engineer"?**

```python
structural_engineers = df[df['profession'] == 'Structural Engineer']
counts = structural_engineers.count()

for column, count in counts.items():
    print(f"{column:15} {count}")
```

**8. How many male customers are 'Structural Engineer'?**

```python
male_structural_engineers  = df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Male')]
counts = male_structural_engineers.count()

for column, count in counts.items():
    print(f"{column:15} {count}")
```

**9. Find out the female Structural Engineers from province Alberta (AB)?**

```python
df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Female') & (df['province'] == 'AB')]
```

**10. What is the max, min and average spending?**

```python
print("Max. spending: ", df['price(CAD)'].max())
print("Min. spending: ", df['price(CAD)'].min())
print("Avg. spending: ", df['price(CAD)'].mean())
```

**11. Who did not spend anything? Company wants to send a deal to encourage the customer to buy stuff!**

**17. What are two most common professions?**

```python
profession = df['profession'].value_counts().head(2)

for prof, count in profession.items():
    print(f"{prof:22} {count}")
```

**18. Can you tell the top 5 most popular email providers? (e.g. gmail.com, yahoo.com, etc...)**

```python
domain = df['email'].str.split('@').str[1]
provider = domain.value_counts().head()

for prov, count in provider.items():
    print(f"{prov:15} {count}")
```

**19. Is there any customer who is using email with "am.edu"?**<br>
Hint: Use `lambda` expression in `apply()`. split the email address at `@`.  

```python
domain = df['email'].str.split('@').str[1]
df[domain=='am.edu']
```

```python
# (lambda parameter : expression)(value pass)
df[df['email'].apply(lambda x: 'am.edu' in x)]
```

```python
# (lambda parameter : expression)(value pass)
df[df['email'].apply(lambda x: x.split('@')[1] == 'am.edu')]
```

**20. Which day of the week, the store gets more customers?**

```python
weekday = df['weekday'].value_counts()

for week, count in weekday.items():
  print(f"{week:15} {count}")
```
