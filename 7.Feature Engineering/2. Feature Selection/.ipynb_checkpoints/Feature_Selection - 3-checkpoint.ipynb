### Corelation Matrix with Heat Map

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cardio_train.csv')
df
```

```python
df = df.drop('id', axis=1)
df
```

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(corr, annot = True, cmap='coolwarm')
```
